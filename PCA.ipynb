In Principal Component Analysis (PCA), **variance** refers to the amount of variability or spread in the data along a particular direction or axis.

### Here's how it works:

1. **Data Representation:** When you have a dataset with multiple features, PCA aims to reduce its dimensionality while retaining as much information as possible. This is done by identifying new axes (called principal components) that maximize the variance in the data.

2. **Principal Components:** The first principal component is the direction in the feature space along which the data has the highest variance. The second principal component is the direction orthogonal to the first component that captures the next highest variance, and so on.

3. **Variance Explained:** Each principal component explains a certain percentage of the total variance in the data. The sum of the variances of all principal components equals the total variance in the original data.

4. **Why Variance Matters:** Higher variance along a principal component means that it captures more of the underlying structure of the data, which is why PCA prioritizes components with higher variance.

### In summary
Variance in PCA is a measure of how much information or variability in the data is captured by each principal component. PCA seeks to maximize this variance in the principal components to ensure that the reduced-dimensional representation retains as much of the original data's information as possible.


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
##Load the Dataset
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dataset = load_breast_cancer()
```

```python
cancer_dataset.keys()
```

```python
print(cancer_dataset.DESCR)
```

```python
df = pd.DataFrame(cancer_dataset['data'], columns=cancer_dataset['feature_names'])
df.head()
```

```python
## Standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data = scaler.transform(df)
```

```python
scaled_data
```

```python
## Applying PCA Algortihms
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
data_pca = pca.fit_transform(scaled_data)
```

```python
data_pca
```

```python
pca.explained_variance_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data_pca[:, 0], data_pca[:,1], c = cancer_dataset['target'], cmap='plasma')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.show()
```
